3. What are the main differences between lists and tuples in Python?
     
     Ans3->> The difference between lists and tuples is lists are mutable , meaning their elements can be changed, added, or removed after the list is created whereas, tuples are immutable, so once a tuple is created, its values cannot be modified. Lists are written using square brackets [ ], while tuples use parentheses ( ). Because tuples cannot be changed, they are generally faster and use less memory compared to lists, Lists also support many built-in methods like append(), remove(), and sort(), whereas tuples have only a few methods such as count() and index(). Due to these differences, lists are used when data needs to be updated frequently, while tuples are preferred for storing fixed and constant data.




4. Describe how dictionaries store data?

   Ans4->>Dictionaries in Python store data in the form of key-value pairs, where each value is accessed using a unique key instead of an index. Internally, dictionaries use a data structure called a hash table, which maps each key to a specific location in memory using a hash function. This allows Python to retrieve, insert, or update values very quickly, usually in constant time. Each key in a dictionary must be unique and immutable, while the values can be of any data type and can be changed as needed. This method of storing data makes dictionaries highly efficient for tasks that require fast lookups, such as searching for information, counting occurrences, or representing structured data.
