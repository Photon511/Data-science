    ans-1>>. Python is a high-level programming language that is very easy to learn and read. It is popular because its syntax is simple, it has many built-in libraries, and it is used in many fields like web development, data science, AI, automation, and more. 2. What is an interpreter in Python?

    ans2>>. An interpreter is a program that reads and runs Python code line by line. Instead of converting the whole program at once, it executes one line at a time. This makes Python easy to test and debug.

3. What are pre-defined keywords in Python?

    ans3>>. Pre-defined keywords are special words in Python that have a fixed meaning. They are already reserved by the language, like if, for, while, True, and False. We use them to write the structure of a program.

4. Can keywords be used as variable names?

    ans4>>. No, keywords cannot be used as variable names because they are already reserved for specific purposes in Python. If we try to use them, Python will show an error.

5. What is mutability in Python?

    ans5>>. Mutability means whether a value can be changed after it is created. If something is mutable, we can modify it. If it is immutable, we cannot change its value once it is made.

6. Why are lists mutable, but tuples are immutable?

    ans6>>. Lists are mutable because Python allows us to add, remove, or change items inside them. Tuples are immutable because once they are created, their values cannot be changed. This helps protect data when we don’t want it to be modified.

7. What is the difference between “==” and “is” operators in Python?

    ans7>>. == checks if the values of two things are the same.
    is checks if both things refer to the exact same memory location.
    So, == compares values, while is compares identity.

8. What are logical operators in Python?

    ans8>>. Logical operators are used to combine conditions. The three main logical operators are and, or, and not. They help us write more complex decision-making statements in programs.

9. What is type casting in Python?

    ans9>>. Type casting means converting one data type into another. For example, changing a string "5" into an integer 5. It helps us use values in the form we need.

10. What is the difference between implicit and explicit type casting?

    ans10>>. Implicit casting happens automatically when Python changes the type by itself.
    Explicit casting happens when the programmer manually converts the type using functions like int(), float(), or str().

11. What is the purpose of conditional statements in Python?

    ans11>>. Conditional statements help the program make decisions. They allow the program to run different code depending on whether a condition is true or false.

12. How does the elif statement work?

    ans12>>. elif stands for “else if.” It is used when there are multiple conditions to check. If the first if condition is false, Python checks the elif condition next.

13. What is the difference between for and while loops?

    ans13>>.  A for loop runs a fixed number of times, usually when we know how many repetitions we need. A while loop runs until a condition becomes false, so the number of repetitions is not always known.

14. Describe a scenario where a while loop is more suitable than a for loop.

    ans14>>. A while loop is better when we don’t know how many times the loop should run. For example, asking a user to enter a password until they get it right. We can't predict how many tries they will need, so a while loop works best.

```python

##14. Write a program to reverse a string using a while loop.
s = input()
i = len(s) - 1
rev = ""

while i >= 0:
    rev += s[i]
    i -= 1

print(rev)
```

```python
##15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
n = int(input())
fact = 1
i = 1

while i <= n:
    fact *= i
    i += 1

print(fact)
```
